```python
import matplotlib.pyplot as plt
import numpy as np

bakterier = 1.e6
sur = 1e7
faktor = 1.5
tid = 0

y_b = [bakterier]


while bakterier < sur:
    bakterier = faktor * bakterier
    y_b.append(bakterier)
    tid += 1
    
x_t = np.arange(tid + 1)

print(f"Det tar {tid} timmar i rumstemporatur för att mjölken ska surna")

print(f" x = {x_t}")
print(f" y = {y_b}")

print(" \n ")

plt.plot(x_t, y_b)
plt.xlabel("Timmar (h)")
plt.ylabel("Antal Bakterier")
plt.title("Bakterietillväxt i mjölk per timme")
plt.show()
```
