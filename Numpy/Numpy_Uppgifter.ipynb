# Uppgift 3

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11)


def f(x):
    y = 3*x + 1
    return y


plt.plot(x,f(x))
plt.title("Formel = 3x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Uppgift 4

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)


def f(x):
    y = x**2 - 1
    return y


plt.plot(x,f(x))
plt.title("Formel = x2 - 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Uppgift 5

```python
import numpy as np
import matplotlib.pyplot as plt


nollstalle1 = -0.04138126
nolstalle2 = 6.04138126

t = np.linspace(nollstalle1,nolstalle2)


def f(t):
    y = (-4)*t**2 + 24*t + 1
    return y


function = (-4)*t**2 + 24*t + 1
print(f"Max Höjden = {np.amax(function)}")

plt.plot(t, f(t))
plt.plot(3,np.amax(function),"ro")
plt.title("Raket Åkbana")
plt.ylabel("Höjd")
plt.xlabel("Tid (s)")
plt.show()
```
