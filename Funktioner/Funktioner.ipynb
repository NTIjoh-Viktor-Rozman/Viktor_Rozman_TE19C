```python
import numpy as np
import matplotlib.pyplot as plt

def N(t):
    return 11/(1+3.4*np.exp(-0.03*t))

print(f"Antal Människor år 1950: {N(0)} miljareder")

år = 400

t = np.arange(år)

plt.plot(t, N(t))
plt.ylabel("Antal människor i miljarder")
plt.xlabel("År sedan 1950")
plt.title("Modell för Population")

print(f"Övre gräns: {N(400):.2f} miljarder")
print(" \n ")
```
