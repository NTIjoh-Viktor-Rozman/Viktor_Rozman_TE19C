```python
# Uppgift 1 #
# A #

import random 

lista = []

for i in range(10):
    kast = random.randint(1,6)
    lista.append(kast)

lista.sort()
print(f"Du kastade en trärning 10 gånger och fick: \n {lista}")
```

```python
# Uppgift 1 #
# B #

import random 

lista = []

for i in range(10):
    kast = random.randint(1,6)
    lista.append(kast)

lista.sort(reverse = True)
print(f"Du kastade en trärning 10 gånger och fick: \n {lista}")
```

```python
# Uppgift 3 #

bokstav = ["A", "B", "C", "D", "E", "F", "G", "H"]
siffror = ["1", "2", "3", "4", "5", "6", "7", "8"]

for i in bokstav:
    for j in siffror:
        ruta = i + j
        print(ruta, end = " ")
    print()
```
