```python
tal = [1,2,3,4,5]
kvadraten = [x**2 for x in tal]

print(kvadraten)
```

```python
frukter = ["äpplen", "påron", "kiwi", "banan", "jordgubbe", "blåbär"]

print(f"Fuktlista: {frukter}")
print(f"Frukter[0]: {frukter[0]}")
print(f"Frukter[-1]: {frukter[-1]}")
print(f"Frukter[3]: {frukter[3]}")
print(f"Frukter[::-1]: {frukter[::-1]}")

print("\n")

for frukt in frukter:
    print(f"Frukt: {frukt}")
```

```python
grönsaker = ["tomat", "gurka", "majs", "sallad"]
frukter = ["äpple", "påron", "kiwi", "banan", "jordgubbe", "blåbär"]

fruktsallad = []

for grönsak in grönsaker:
    fruktsallad.append(grönsak)

for frukt in frukter:
    fruktsallad.append(frukt)

print(fruktsallad)
```

```python
import matplotlib.pyplot as plt

y = [2*x-2 for x in range(10)]
x = [x for x in range(10)]

print(f"x: {x}")
print(f"y: {y}")

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graf till Y = 2*x-2")
```

```python
tabell = []

for i in range(11):
    rad = [x*i for x in range(11)]
    tabell.append(rad)
    for j in range(11):
        print(f"{tabell[i][j]:3}", end = " ")

    print()

print()

print(tabell)
print("\n")
print(f"Tabell[3]: {tabell[3]}")
print(f"Tabell[3][3]: {tabell[3][3]}")
```
