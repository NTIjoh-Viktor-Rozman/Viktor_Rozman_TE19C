```python
## Uppgift E ##

import math
import random
import matplotlib.pyplot as plt

Antal = 5000 ## Använder en satt mängt punkter i 'for' satsen så att jag kan senarekan använda dem                    för beräkningar ##
num = 0

for i in range(Antal):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    d = math.sqrt(x**2 + y**2)
    if d < 1:
        plt.plot(x, y,'*r')
        num += 1 ## Ett nummer som säger hur många som träffade, behövs för beräkning ##
    else:
        plt.plot(x, y,'*b') ## Allt annat missar, därför kan man använda 'else' ##

## Beräknar om punkten är innanför cirklen med pitagoras sats ##

Decimal = num / Antal
Procent = Decimal * 100 

## Räknar ut procenten som träffade ##

print(f"Du kastade {Antal} gånger och träffade {num} gånger, det betyder att du träffade {Procent:.5f}% av Kasten.")

print(" \n ") ## Mellansteg mellan text och graf ##
```

```python
## Uppgift G ##

import math
import random
import matplotlib.pyplot as plt

Antal = 10000

for i in range(-1, Antal):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    d = math.sqrt(x**2 + y**2)
    if d < 1:
        plt.plot(x, y,'*r') ## '*r' betyder att pricken ska vara en stjärna och röd ##
    else:
        plt.plot(x, y,'.b') ## '.b' betyder att pricken ska vara en punkt och blå ##

## Då jag redan råkat göra detta på en tidigare uppgift gjorde jag också så träffarna utanför cirklen     visades med en prick isstället för en stjärna ##
```

```python
## Uppgift E ##

Start = int(input("Ange ett tal där du vill att din nummer linje ska starta: "))
Slut = int(input("Ange ett tal där du vill att nin nummer linje ska sluta: "))
Slut2 = Slut + 1 ## Eftersom om du vill att den ska sluta på 100 kommer den bara rabla upp till 99 ##

num = float(input("Ange ett tal som ska ersättas med Burr: "))
num2 = float(input("Ange ett till tal som ska ersättas med Birr: "))

Burr = 0
Birr = 0
BurrBirr = 0

## Angera att alla variabler är 0 från början ##

for i in range (Start,Slut2):
    if i %num == 0 and i %num2 == 0:
        print("Burr Birr", end = ' ') ## Lägger till BurrBirr ##
        BurrBirr += 1 ## Berättar för datorn att den ska lägga till en på BurrBirr ##
    elif i %num == 0:
        print("Burr", end = ' ')
        Burr += 1 ## Berättar för datorn att den ska lägga till en på Burr ##
    elif i %num2 == 0:
        print("Birr", end = ' ')
        Birr += 1 ## Berättar för datorn att den ska lägga till en på Birr ##
    else:
        print(i, end = ' ')

print(" \n ")

print(f"Det finns {BurrBirr} stycken BurrBirr i din tallinje")
print(f"Det finns {Burr} stycken Burr i din tallinje")
print(f"Det finns {Birr} stycken Birr i din tallinje")

## Printar hur många av Burr, Birr och BurrBirr som fanns med i tallinjen ##
```
