```python
## Uppgift E ##

import math
import random
import matplotlib.pyplot as plt

Antal = 5000
num = 0

for i in range(Antal):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    d = math.sqrt(x**2 + y**2)
    if d < 1:
        plt.plot(x, y,'*r')
        num += 1
    else:
        plt.plot(x, y,'*b')

Decimal = num / Antal
Procent = Decimal * 100 

print(f"Du kastade {Antal} gånger och träffade {num} gånger, det betyder att du träffade {Procent:.5f}% av Kasten.")

print(" \n ")
```

```python
## Uppgift D ##

num = float(input("Ange ett tal: "))
num2 = float(input("Ange ett till tal: "))

for i in range (1,101):
    if i %num == 0:
        print("Burr", end = ' ')
    elif i %num2 == 0:
        print("Birr", end = ' ')
    else:
        print(i, end = ' ')
```

```python
## Uppgift E ##

Start = int(input("Ange ett tal där du vill att din nummer linje ska starta: "))
Slut = int(input("Ange ett tal där du vill att nin nummer linje ska sluta: "))
Slut2 = Slut + 1

num = float(input("Ange ett tal som ska ersättas med Burr: "))
num2 = float(input("Ange ett till tal som ska ersättas med Birr: "))

Burr = 0
Birr = 0
BurrBirr = 0

for i in range (Start,Slut2):
    if i %num == 0 and i %num2 == 0:
        print("Burr Birr", end = ' ')
        BurrBirr += 1
    elif i %num == 0:
        print("Burr", end = ' ')
        Burr += 1
    elif i %num2 == 0:
        print("Birr", end = ' ')
        Birr += 1
    else:
        print(i, end = ' ')

print(" \n ")

print(f"Det finns {BurrBirr} stycken BurrBirr i din tallinje")
print(f"Det finns {Burr} stycken Burr i din tallinje")
print(f"Det finns {Birr} stycken Birr i din tallinje")
```
