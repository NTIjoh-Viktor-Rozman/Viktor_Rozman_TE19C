```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.recsv("sysselsattning.csv", header=2)

Nej = df[df["sysselsättning"] == "ej förvärvsarbetande"]
NejSum = Nej["2019"].sum()

Ja = df[df["sysselsättning"] == "förvärvsarbetande"]
JaSum = Ja["2019"].sum()

sizes = [JaSum, NejSum]
labels = ["förvärvsarbetande", "ej förvärvsarbetande"]

explode = [0.05, 0]

plt.pie(sizes, explode=explode, labels=labels)
plt.show()
```

```python
import pandas as pd 

df = pd.read_csv("sysselsattning.csv", header=2)
ja = df[df["sysselsättning"] == "förvärvsarbetande"]
nej = df[df["sysselsättning"] == "ej förvärvsarbetande"]

mänJa = ja[ja["kön"] == "män"]
mänJa = mänJa["2019"].sum()
mänNej = nej[nej["kön"] == "män"]
mänNej = mänNej["2019"].sum()

kvinnorJa = ja[ja["kön"] == "kvinnor"]
kvinnorJa = kvinnorJa["2019"].sum()
kvinnorNej = nej[nej["kön"] == "kvinnor"]
kvinnorNej = kvinnorNej["2019"].sum()

sizes = [mänJa, mänNej, kvinnorJa, kvinnorNej]
labels = ["Män (JA)", "Män (NEJ)", "Kvinnor (JA)", "Kvinnor (NEJ)"]

plt.bar(labels, sizes)
plt.title("BLEB!")
plt.show()
```

 # (3) Gör en Graf som visar på hur andelen förvärvsarbetande förändras för män genom åldersgrupperna. Gör ytterligare en graf som visar på andelen förvärvsarbetande förändras för kvinnor genom åldersgrupperna.

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv("sysselsattning.csv", header=2)
män = df[df["kön"] == "män"]
mänJa = män[män["sysselsättning"] == "förvärvsarbetande"]
ListaAntalMän = mänJa["2019"]
ListaÅlderMän = mänJa["ålder"]

kvinnor = df[df["kön"] == "kvinnor"]
kvinnorJa = kvinnor[kvinnor["sysselsättning"] == "förvärvsarbetande"]
ListaAntalKvinnor = kvinnorJa["2019"]
ListaÅlderKvinnor = kvinnorJa["ålder"]

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(ListaÅlderKvinnor, ListaAntalKvinnor)
plt.title("Antalet Kvinnor som Förvarvsarbetar 2019")

plt.subplot(1,2,2)
plt.plot(ListaÅlderMän, ListaAntalMän)
plt.title("Antalet Män som Förvarvsarbetar 2019")
plt.show()
```
